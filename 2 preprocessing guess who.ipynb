```python
import pandas as pd
df = pd.DataFrame.from_csv('parsed_json/guess-who-data-2.csv', index_col = None)
```

```python
df1 = df.dropna(subset=['question', 'eliminatedFaces'])
```

```python
df1["question"]
```

```python
def cleanQuestion(): 
    questions = df1["question"]
    cleanQuestions = []
    for question in questions:
        NewQuestion = question.replace("[comma]", "").replace("'", "").strip("?,''.").lower()
        cleanQuestions.append(NewQuestion)
    return cleanQuestions
        
```

```python
questions = cleanQuestion()
questions
```

```python
def tokenize(questionlist):
    tokens = []
    for sentence in questionlist:
        tokens.append(sentence.split())
    return tokens
        
    
```

```python
tokens = tokenize(questions)
```

```python
import gensim
import os
```

```python
model = gensim.models.Word2Vec(tokens, size=150, window=10, min_count=2, workers=10)
model.train(tokens, total_examples=len(tokens), epochs=10)
```

```python
w1 = "face"
print("Most similar to {0}".format(w1), model.wv.most_similar(positive=w1))
```

```python
import re
def cleanQuestion(): 
    questions = df['question']
    cleanQuestions = []
    for question in questions:
        NewQuestion = question.replace("[comma]", "").lower()
        NewQuestion = re.sub("[^A-Za-z0-9 ]*", "", NewQuestion)
        NewQuestion = NewQuestion.strip()
        cleanQuestions.append(NewQuestion)
    return cleanQuestions
questions = cleanQuestion()
```
