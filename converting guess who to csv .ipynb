```python
import json
import csv

json_file = open("parsed_json/guess-who-data-1.json")
json_data = json.loads(json_file.read())
json_file.close()

csv_file = open("parsed_json/guess-who-data-1.csv", "w")

headerWritten = False
for action in json_data:
    row = {}
    row['allFaces']        = ','.join(action['allFaces'])
    row['eliminatedFaces'] = ','.join(action['eliminatedFaces'])

    row['question'] = action['question']
    row['questionResponse']  = action['questionResponse']
    row['uniqueID'] = action['uniqueID']
    
    columns    = sorted(row.keys())
    csv_writer = csv.DictWriter(csv_file, columns)

    if not headerWritten:
        csv_writer.writeheader()
        headerWritten = True

    csv_writer.writerow(row)

csv_file.close()

print('done!')
```
