```python
import pandas as pd
data = pd.read_csv('processed-data.csv')
df = data[['faceId', 'question','gender', 'ethnicity','hair', 'eyes', 'trueAnswer']]
```

```python
df.dtypes
```

```python
#pre-processing for logistic regression
col_list = ["gender", 'ethnicity', 'hair', 'eyes', 'trueAnswer']
obj_df = df.loc[:,col_list]
obj_df = obj_df.astype('category')
obj_df['trueAnswer'] = obj_df['trueAnswer'].cat.codes
obj_df['gender'] = obj_df['gender'].cat.codes
obj_df['hair'] = obj_df['hair'].cat.codes
obj_df['eyes'] = obj_df['eyes'].cat.codes
obj_df['ethnicity'] = obj_df['ethnicity'].cat.codes
obj_df.to_csv('check-data.csv', index = False)
```

```python
y = obj_df["trueAnswer"]
X = obj_df.loc[:,['hair', 'eyes', 'ethnicity', 'gender']]
```

```python
#train set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=0.2, random_state= 0)
```

```python
from sklearn.linear_model import LogisticRegression 
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_test = lr.predict(X_test)
```
