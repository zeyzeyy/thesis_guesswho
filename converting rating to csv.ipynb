```python
import json
import csv

json_file = open("parsed_json/face-rating-data-2.json")
json_data = json.loads(json_file.read());
json_file.close();

csv_file = open("parsed_json/face-rating-data-2.csv", "w")

headerWritten = False
for action in json_data:
    
    action['faceID'] = action['faceID']

    action['eye']  = action['responses']['eye']
    action['hair'] = action['responses']['hair']
    action['age'] = action['responses']['age']
    action['gender'] = action['responses']['photo-gender']
    if 'ethnicity' in action['responses']:
        action['ethnicity'] = action['responses']['ethnicity']
    else: 
        action['ethnicity'] = None

    action.pop('experimentCode')
    action.pop('completion_code')
    action.pop('uniqueID')
    action.pop('demographics')
    action.pop('state')
    action.pop('trialNumber')
    action.pop('responses')
    
    columns    = sorted(action.keys())
    csv_writer = csv.DictWriter(csv_file, columns)

    if not headerWritten:
        csv_writer.writeheader()
        headerWritten = True

    csv_writer.writerow(action);

csv_file.close();

print('done!')
```
